```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
df1 = pd.read_csv('sales_data.csv')
df2 = pd.read_csv('customer_data.csv')
df3 = pd.read_csv('product_data.csv')
df4 = pd.read_csv('store_data.csv')
```

#Join DF

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.merge(df1, df2, on='customer_id', how='left')
```

```python
df.head()
```

```python
df3.head()
```

```python
df_join = pd.merge(df, df3, on='product_id', how='left')
```

```python
df_join.head()
```

```python
df_final = pd.merge(df_join, df4, on='store_id', how='left')
```

```python
df_final.head()
```

#Data Cleaning dan Manipulation

```python
df_final.info()
```

##duplicate

```python
cek_duplicate = df_final.duplicated().sum()

print(cek_duplicate)
```

##missing value

```python
df_final.isna().sum()
```

```python
# Hitung persentase missing value

total_data = len(df_final)
total_missing = df_final.isna().sum()
total_missing_percentage = (total_missing / total_data) * 100

print(total_missing_percentage)
```

```python
# 1. Cek total data awal
print(f"Total data awal: {len(df_final)}")
print(f"Total customers awal: {df_final['customer_id'].nunique()}")

# 2. Drop rows dengan product_id = P999999 (data tidak lengkap)
df_clean = df_final[df_final['product_id'] != 'P999999'].copy()
print(f"\nSetelah drop product P999999: {len(df_clean)} rows")

# 3. Drop rows tanpa customer_id (untuk RFM analysis)
df_clean = df_clean[df_clean['customer_id'].notna()].copy()
print(f"Setelah drop missing customer_id: {len(df_clean)} rows")
print(f"Total customers: {df_clean['customer_id'].nunique()}")

# 3b. Drop rows dengan store_name atau region kosong
before = len(df_clean)
df_clean = df_clean.dropna(subset=['store_name', 'region'])
after = len(df_clean)
print(f"\nSetelah drop missing store_name/region: {after} rows (dihapus {before - after})")

# 4. Handle discount - isi dengan 0 asumsi tidak ada diskon
df_clean['discount'] = df_clean['discount'].fillna(0)

# 5. Handle color - isi dengan 'Unknown', karena ga penting dalam analisis
df_clean['color'] = df_clean['color'].fillna('Unknown')

# 6. Cek missing values terakhir
print("\nMissing values setelah cleaning:")
print(df_clean.isnull().sum())

# 7. Summary
print(f"\n{'='*50}")
print(f"DATA CLEANING SUMMARY")
print(f"{'='*50}")
print(f"Data retained: {len(df_clean)/len(df_final)*100:.1f}%")
print(f"Rows dropped: {len(df_final) - len(df_clean)}")
print(f"Customers untuk RFM: {df_clean['customer_id'].nunique()}")
```

##data manipulation

```python
#drop kolom yang tidak penting untuk analisis agar tidak berat
df_clean = df_clean.drop(['store_id', 'email', 'size', 'color', 'season', 'supplier', 'store_size_m2'], axis=1)
```

```python
df_clean['age'] = df_clean['age'].astype(int)
df_clean['date'] = pd.to_datetime(df_clean['date'])
```

```python
#buat kolom sales
df_clean['sales'] = df_clean['list_price'] * df_clean['quantity'] * (1 - df_clean['discount'])
```

```python
#buat kolom profit
df_clean['profit'] = df_clean['sales'] - (df_clean['cost_price'] * df_clean['quantity'])
```

```python
#hapus transaksi returned dari data
df_clean = df_clean[df_clean['returned'] == 0]
```

```python
df_clean = df_clean.drop('returned', axis=1)
```

```python
plt.figure(figsize=(12,6))

# plotting
features = ['quantity','discount','age','cost_price','list_price','sales','profit']
for i in range(0, len(features)):
    plt.subplot(1, len(features), i+1)
    sns.boxplot(y=df_clean[features[i]], color='red')
    plt.tight_layout()
```

Outlier di sales atau profit bisa jadi pelanggan bernilai tinggi (high-value customers) yang penting untuk analisis. Menghapusnya bisa membuat kita kehilangan insight tentang pelanggan yang paling berkontribusi besar. profit negatif juga wajar karena artinya transaksi tersebut merugikan (biaya lebih besar dari penjualan), bisa jadi karena efek diskon.
